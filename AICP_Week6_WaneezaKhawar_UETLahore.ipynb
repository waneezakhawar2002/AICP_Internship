```python
# TASK 1
def check_sack(sack, weight):    
    if (sack=='S'):
        if (weight>=49.9 and weight<=50.1):
            return 0
        elif (weight<49.9):
            return -1
        elif (weight>50.1):
            return 1
        
    elif (sack=='G'):
        if (weight>=49.9 and weight<=50.1):
            return 0
        elif (weight<49.9):
            return -1
        elif (weight>50.1):
            return 1
        
    elif (sack=='C'):
        if (weight>=24.9 and weight<=25.1):
            return 0
        elif (weight<24.9):
            return -1
        elif (weight>25.1):
            return 1
    else:
        return 2




while (True):
    print("Enter C for cement, G for gravel, S for sand.")
    print("What is in the sack?")
    sack = input()
    sack = sack.upper()
    if (sack=='S'):
        sack_content = 'Sand'
    elif (sack=='C'):
        sack_content = 'Cement'
    elif (sack=='G'):
        sack_content = 'Gravel'
        
    print("What is the weight of the sack?")
    weight = float(input())
    valid = check_sack(sack,weight)
    if (valid == 2):
        print("Sack Rejected!\nThe sack has unknown contents.")
    elif (valid == -1):
        print(f"Sack Rejected!\nThe sack contains {sack_content} but is underweight.")
    elif (valid == 1):
        print(f"Sack Rejected!\nThe sack contains {sack_content} but is overweight.")
    if (valid == 0):
        print(f"Sack Accepted!\nThe sack contains {sack_content} and weighs {weight} kg.")
    
    more = input("Want more sacks? (N/Y)\n")
    more = more.upper()
    if (more=='N'):
        break
    
    
```

```python
# TASK 2

rejected_sacks = 0
total_weight = 0

num1 = int(input("How many sacks of Cement do you want?"))
for i in range (0,num1):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('C',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_sacks = rejected_sacks + 1
        
num2 = int(input("\nHow many sacks of Gravel do you want?"))
for i in range (0,num2):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('G',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_sacks = rejected_sacks + 1
        
num3 = int(input("\nHow many sacks of Sand do you want?"))
for i in range (0,num3):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('S',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_sacks = rejected_sacks + 1
        
print(f"\nTotal weight of the order is {total_weight} kg.")
print(f"\nThe number of sacks rejected from the order is {rejected_sacks}")
```

```python
# TASK 3
total_weight = 0
regular_price = 0

num1 = int(input("How many sacks of Cement do you want?"))
cement_sacks = 0
rejected_csacks = 0
for i in range (0,num1):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('C',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_csacks = rejected_csacks + 1
cement_sacks = num1 - rejected_csacks
regular_price = regular_price + 3*cement_sacks
        
num2 = int(input("\nHow many sacks of Gravel do you want?"))
gravel_sacks = 0
rejected_gsacks = 0
for i in range (0,num2):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('G',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_gsacks = rejected_gsacks + 1
gravel_sacks = num2 - rejected_gsacks  
regular_price = regular_price + 2*gravel_sacks
        
num3 = int(input("\nHow many sacks of Sand do you want?"))
sand_sacks = 0
rejected_ssacks = 0
for i in range (0,num3):
    w = float(input(f"Enter the weight of sack no. {i+1}: "))
    valid = check_sack('S',w)
    if (valid==0):
        total_weight = total_weight + w
    else:
        rejected_ssacks = rejected_ssacks + 1
sand_sacks = num3 - rejected_ssacks
regular_price = regular_price + 2*sand_sacks
        
rejected_sacks = rejected_csacks + rejected_gsacks + rejected_ssacks
    
print(f"\nTotal weight of the order is {total_weight} kg.")
print(f"\nThe number of sacks rejected from the order is {rejected_sacks}")
print(f"\nThe regular price of the order is ${regular_price}")

# Finding Special Packs
special_packs = 0
while (cement_sacks>=1 and gravel_sacks>=2 and sand_sacks>=2):
    cement_sacks = cement_sacks-1
    gravel_sacks = gravel_sacks-2
    sand_sacks = sand_sacks-2
    special_packs = special_packs+1
    
if (special_packs>0):
    new_price = 10*special_packs+3*cement_sacks+2*gravel_sacks+2*sand_sacks
    print(f"\nThere are {special_packs} special packs in the order.")
    print(f"So the new price is ${new_price}")
```
