```python
item_code = ['A1','A2','B1','B2','B3','C1','C2','C3','D1','D2','E1','E2','E3','F1','F2','G1','G2']
description = ['Compact','Tower  ','8GB','16GB','32GB','1TB HDD','2TB HDD','4TB HDD','240GB SSD','480GB SSD','1TB HDD','2TB HDD','4TB HDD','DVD/Blu-Ray Player','DVD/Blu-Ray Re-writer','Standard Version','Professional Version']
price = [75.00,150.00,79.99,149.99,299.99,49.99,89.99,129.99,59.99,119.99,49.99,89.99,129.99,50.00,100.00,100.00,175.00]
#print(len(item_code),len(description),len(price))
```

```python
""" TAKS 1 """
total_price = 200
print("The available Cases are:")
print("Item Code\tDescription\tPrice($)")
for i in range (2):
    print(item_code[i],'\t\t',description[i],'\t',price[i]) 
case = input("Choose a Case ITEM CODE:")
case = case.upper()
print("Case:",case)
if (case!='A1' and case!='A2'):
    raise Exception("Selected Case Item Code is not valid!")
for i in range (17):
    if (item_code[i]==case):
        index = i
total_price = total_price+price[index]
    
print("\nThe available RAMs are:")
print("Item Code\tDescription\tPrice($)")
for i in range (2,5):
    print(item_code[i],'\t\t',description[i],'\t',price[i]) 
RAM = input("Choose a RAM ITEM CODE:")
RAM = RAM.upper()
print("RAM:",RAM)
if (RAM!='B1' and RAM!='B2' and RAM!='B3'):
    raise Exception("Selected RAM Item Code is not valid!")
for i in range (17):
    if (item_code[i]==RAM):
        index = i
total_price = total_price+price[index]

print("\nThe available Main Hard Disk Drives are:")
print("Item Code\tDescription\tPrice($)")
for i in range (5,8):
    print(item_code[i],'\t\t',description[i],'\t',price[i]) 
Hard_Disk = input("Choose a Main Hard Disk Drive ITEM CODE:")
Hard_Disk = Hard_Disk.upper()
print("Main Hard Disk Drive:",Hard_Disk)
if (Hard_Disk!='C1' and Hard_Disk!='C2' and Hard_Disk!='C3'):
    raise Exception("Selected Main Hard Disk Drive Item Code is not valid!")
for i in range (17):
    if (item_code[i]==Hard_Disk):
        index = i
total_price = total_price+price[index]

print("The chosen items are:\nCase:",case,"\nRAM:",RAM,"\nMain Hard Disk Drive:",Hard_Disk)
print(f"The Total Price of the Computer is ${total_price}")
```

```python
""" TASK 2 """
print("Other Categories:\nSolid State Drive\nSecond Hard Disk Drive\nOptical Drive\nOperating System")
Other_items = input("Do you want to purchase any items from the other categories (YES or NO):")
Other_items = Other_items.upper()
if (Other_items == 'NO'):
    print("Thank you!")
elif (Other_items == 'YES'):
    No_of_other_items=int(input("Tell us how many items do you want:"))
    print("Now write the key of the item you want.")
    print("Keys of items are as follows:")
    print("Solid State Drive:\t0")
    print("Second Hard Disk Drive:\t1")
    print("Optical Drive:\t\t2")
    print("Operating System:\t3")
    Other_categories = []
    Chosen_items = []
    for i in range (No_of_other_items):
        temp = int(input(f"Key of Item No. {i+1}: "))
        if (temp!=0 and temp!=1 and temp!=2 and temp!=3):
            raise Exception("The provided key is not valid!")
        Other_categories.append(temp)
    for j in range (len(Other_categories)):
        if (Other_categories[j] == 0):
            pop_index = j
            print("The available Solid State Drives are:")
            print("Item Code\tDescription\tPrice($)")
            for i in range (8,10):
                print(item_code[i],'\t\t',description[i],'\t',price[i]) 
            SSD = input("Choose a Solid State Drive ITEM CODE:")
            SSD = SSD.upper()
            print("Solid State Drive:",SSD)
            if (SSD!='D1' and SSD!='D2'):
                raise Exception("Selected Solid State Drive Item Code is not valid!")
            for i in range (17):
                if (item_code[i]==SSD):
                    index = i
            total_price = total_price+price[index]
            print(total_price)
            Chosen_items.append(f"Solid State Drive: {SSD}")
            #Other_categories.pop(pop_index)
        if (Other_categories[j] == 1):
            pop_index = j
            print("The available Second Hard Disk Drives are:")
            print("Item Code\tDescription\tPrice($)")
            for i in range (10,13):
                print(item_code[i],'\t\t',description[i],'\t',price[i]) 
            Second_HDD = input("Choose a Second Hard Disk Drive ITEM CODE:")
            Second_HDD = Second_HDD.upper()
            print("Second Hard Disk Drive:",Second_HDD)
            if (Second_HDD!='E1' and Second_HDD!='E2' and Second_HDD!='E3'):
                raise Exception("Selected Second Hard Disk Drive Item Code is not valid!")
            for i in range (17):
                if (item_code[i]==Second_HDD):
                    index = i
            total_price = total_price+price[index]
            print(total_price)
            Chosen_items.append(f"Second Hard Disk Drive: {Second_HDD}")
            #Other_categories.pop(pop_index)
        if (Other_categories[j] == 2):
            pop_index = j
            print("The available Optical Drives are:")
            print("Item Code\tDescription\tPrice($)")
            for i in range (13,15):
                print(item_code[i],'\t\t',description[i],'\t',price[i]) 
            Optical_Drive = input("Choose an Optical Drive ITEM CODE:")
            Optical_Drive = Optical_Drive.upper()
            print("Optical Drive:",Optical_Drive)
            if (Optical_Drive!='F1' and Optical_Drive!='F2'):
                raise Exception("Selected Optical Drive Item Code is not valid!")
            for i in range (17):
                if (item_code[i]==Optical_Drive):
                    index = i
            total_price = total_price+price[index]
            print(total_price)
            Chosen_items.append(f"Optical Drive: {Optical_Drive}")
            #Other_categories.pop(pop_index)
        if (Other_categories[j] == 3):
            pop_index = j
            print("The available Operating Systems are:")
            print("Item Code\tDescription\tPrice($)")
            for i in range (15,17):
                print(item_code[i],'\t\t',description[i],'\t',price[i]) 
            Operating_System = input("Choose an Operating System ITEM CODE:")
            Operating_System = Operating_System.upper()
            print("Operating System:",Operating_System)
            if (Operating_System!='G1' and Operating_System!='G2'):
                raise Exception("Selected Operating System Item Code is not valid!")
            for i in range (17):
                if (item_code[i]==Operating_System):
                    index = i
            total_price = total_price+price[index]
            print(total_price)
            Chosen_items.append(f"Operating System: {Operating_System}")
            #Other_categories.pop(pop_index)
    print("The Additional Chosen Items are")
    for i in range (len(Chosen_items)):
        print(Chosen_items[i])
    print(f"The new total price of the Computer is ${total_price}")
else:
    raise Exception("Invalid Answer")
```

```python
""" TASK 3 """
money_saved = 0
new_price = total_price
print(f"Number of Additional Items purchased is {len(Chosen_items)}.")
if (len(Chosen_items)==1):
    money_saved = total_price*0.05
    new_price = total_price-money_saved
elif (len(Chosen_items)>=2):
    money_saved = total_price*0.10
    new_price = total_price-money_saved
print(f"The amount of money saved is ${money_saved}")
print(f"The new price of the Computer is ${new_price}")
```
